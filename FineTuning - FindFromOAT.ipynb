```python
import pandas as pd
import numpy as np
import os
```

```python
username = ""
MODEL_PATHS = f"/home/{username}/.pyenv/runs/detect"
```

```python
for idx,p in enumerate(os.listdir(MODEL_PATHS)):
    name = p
    p = os.path.join(MODEL_PATHS, p)
    if idx == 0:
        og_df = pd.read_csv(os.path.join(p, "results.csv"))
        og_df = og_df.iloc[-3:].mean(axis=0)
        og_df = pd.DataFrame(og_df).T
        og_df.insert(loc=0, column='name', value=name)
    else:
        df = pd.read_csv(os.path.join(p, "results.csv"))
        df = df.iloc[-3:].mean(axis=0)
        df = pd.DataFrame(df).T
        df.insert(loc=0, column='name', value=name)

    
        og_df = pd.concat((og_df, df), axis=0)
og_df.drop_duplicates()
```

```python
og_df.sort_values(by='metrics/mAP50-95(B)', ascending=False).iloc[:10]
```
